## Data Frames: 


This week, we will discover Data Frames--a fundamental data structure to deal with data sets! Data frames are arguably the best method to store information on R.

A data frame is closely related to a list. Essentially, a data frame is a sub category of a list because data frames have multiple restrictions on how they can be created. Thus, it embodies a fusion of a list and a matrix with the class, "data.frame". 

**Summary:**

A data frame may be regarded as a matrix with columns with differing modes and attributes. It may be presented in a matrix-like form, where its rows and columns can be extracted using matrix/list indexing conventions as we saw last week.

A data frame will contain variables and observations. A **variable** will correspond to information in **columns**. An **observation** will correspond to information in **rows**. 
    - This will be revisited later in the tutorial when we talk about structure.


**Restrictions/Requirements for a Data Frame**

1) Components must consist of vectors, factors, numeric matrices, lists, or other data frames.

2) Matrices, lists, and data frames provide as many variables to the new data frame as they have columns, elements, or variables, respectively.

3) Numeric vectors, logicals, and factors are left alone. Character vectors are coerced to be factors, whose levels are the unique values appearing in the vector.

4) If you are adding vector structures to the data frame, it is important to note that *all vector structures* must be the SAME LENGTH! Similarly, *matrix structures* must all have the SAME ROW SIZE. 


**Why not use a matrix for data sets? Isn't this what a data frame basically is?**

Not necessarily! The main reason a data frame is so useful to manipulate datasets is to include **numerics, characters, and other classes of objects.** You may ask why not use a matrix. That's because a matrix does not have the versatility of a data frame in the sense that it can only hold a *single mode*.

Keep in mind that vectors and matrices *cannot* hold more than one type of mode. Lists and data frames solve this issue. 

**NOTE:** Elements in a column must all be the same type. This adds to the organization of a data frame. 


```r
# assign vector with characters and numeric
isla_vista <- c("freebirds", "habit", "ikes", "blaze", 7/11)

# display vector
isla_vista

# display the mode of the vector
# can you guess what the mode will be? explain?
mode(isla_vista)
```

## Creating a Data Frame


Keep in mind that a data frame, in essence, is just an organized data set! Examples of where data sets imported from include CSV files, relational databases, and software packages. 


However, what if you want to create a data frame from scratch using the amazing R programming language??

Use the function:

**REMEMBER:** Keep in mind that one of the special requirements for a data frame is that whatever you input to comprise the data frame *must* all have the same length.

- Following the header, each horizontal row underneath it is called a **data row**.

## Subscripting and Subsetting

- The rules for data frames are very similar to those of a list.

- Subscripting is explicitly identifying an element, using numeric integers, from a data frame.

- Subsetting is assigning that subscripted element(s) to a variable.


## Three Ways to Subscript and Subset:

**If you know the row and column numbers, or names, this method is your best bet. You can also select multiple rows/columns using the ' : '.**
    
<code> 
    dataframe_name[row_number, col_number]
</code>



**If your perspective is to find values given a certain condition (eg >, <, !=, etc.), then this method is ideal.**
    
<code> 
    subset(my_df, subset = some_condition)
</code>



**If you want a whole column's value, and you know the column name, then use this method.**
    
<code>
    data_frame$column_name
</code>




## Examples

```r
forecast <- c("showers", "cloudy", "cloudy", "rain", "rain")
low_temperature <- c(46, 34, 44, 55, 53)
precipitation <- c(TRUE, FALSE, FALSE, TRUE, TRUE)
weather <- data.frame(forecast, low_temperature, precipitation) 
#added row names to make it easier to reference
row.names(weather) <- c("day1", "day2", "day3", "day4", "day5")

# original data frame (reference)
weather

# specific
weather[1, 2]

# conditional 
# (the argument "subset" --> logical expression indicating elements or rows to keep: missing values are taken as false.)
subset(weather, subset = precipitation == 1)

# a whole column
weather$low_temperature
```

## Finding Specific Values Using Conditionals

```r
forecast <- c("showers", "cloudy", "cloudy", "rain", "rain")
low_temperature <- c(46, 34, 44, 55, 53)
precipitation <- c(TRUE, FALSE, FALSE, TRUE, TRUE)
weather <- data.frame(forecast, low_temperature, precipitation) 

# Identify days where the lowest temperature was greater 40
weather[weather$low_temperature > 40, ]

# Identify days where the lowest temperature was greater than the average low temperature
weather[weather$low_temperature > mean(weather$low_temperature), ]

# Identify days where the lowest temperature was greater than the average low temperature AND it was raining
weather[weather$low_temperature > mean(weather$low_temperature) & (weather$precipitation == 1), ] 
```

## Some Useful Functions To Make Data Frames Easier To Deal With

**head(data_frame, n = number of rows returned)**
- head() ==> will return first 'n' rows starting from the beginning of the data frame.


**tail(data_frame, n = number of rows returned)**
- tail() ==> will return first 'n' rows starting from the end of the data frame.


**summary(data_frame)**
- summary() ==> a function that will produce a gist of the results of each individual header. Examples of the statistical summaries that will be outputted include: 
    - Minimum
    - 1st Quartile
    - Median
    - Mean
    - 3rd Quartile
    - Maximum

- However, the function invokes particular methods which depend on the class of the first argument.


**str(data_frame)**
- str() ==> an alternative to 'summary()'. Displays the internal **str**ucture of a data frame, or any other object placed in parentheses.


**glimpse(data_frame)**
- glimpse() ==> basically transposes (the rows become the columns and the columns become the rows) the data frame

- The header becomes vertically-orientated. This function can sometimes help you see the data better becuase of the formatting.


**nrow(data_frame)**
- nrow() ==> gives you the **n**umber of **row**s in the data frame


**ncol(data_frame)**
- ncol() ==> gives you the **n**umber of **col**umns in the data frame


**dim(data_frame)**
- dim() ==> a combination of nrow and ncol. Outputs rows and wolumns respectively.

- **Note** dim(data_frame)[1] == nrow(data_frame)
- **Note** dim(data_frame)[2] == ncol(data_frame)


**colnames(data_frame)**
- colnames() ==> outputs a vector of the header, or the **col**umn **names**


**rownames(data_frame)**
- colnames() ==> outputs a vector of the **row** **names**


**cbind(data_frame)**
- cbind() ==> combines multiple data frames into a single data frame (extends the data frame horizontally)

- Keep in mind that the length of column of the new data frame or vector you are adding must equal the length of the column of the original data frame

    - Check vector with this: 
        - <code> dim(og_df)[1] == length(new_vector) </code>
        - the number of rows need to equal the number of elements
        
    - Check data frame with this:
        - <code> dim(og_df)[1] == dim(new_df)[1] </code>

- separate the each data frame with a comma (e.g.: cbind(x, y, z) )


**rbind(data_frame)**
- rbind() ==> combines multiple data frames into a single data frame (extends the data frame vertically)

- Keep in mind that the length of row of the new data frame or vector you are adding must equal the length of the row of the original data frame

    - Check vector with this: 
        - <code> dim(og_df)[2] == length(new_vector) </code>
        - the number of columns need to equal the number of elements
        
    - Check data frame with this:
        - <code> dim(og_df)[2] == dim(new_df)[2] </code>

- separate the each data frame with a comma (e.g.: rbind(x, y, z) )


**is.na(data_frame)** 
- is.na() ==> returns TRUE if an element in the data frame is missing and FALSE if not


**order(column)**
- order() ==> will help you sort your data by numerically ranking columns (e.g. 1, 2, 3, etc.) in your data frame if that column consists of *only* numbers

- adding a negative sign to your argument would order your numerics in *descending* order. (positive is ascending by default)


**names(data_frame)**
- names() ==> if argument is a data frame, then you will output a vector with the names of the header

- names(data_frame) <- c(list of names)
    - will assign a vector to the header(first row) of the data frame

**fix(data_frame)**
- fix() ==> easy way to access the data editor so you can fix whatever element in the data frame that you want to


**help(data_frame)**
- help() ==> gives you additional information about the given data frame if it is available in R documentation
